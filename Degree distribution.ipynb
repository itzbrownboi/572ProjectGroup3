```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

#Read Data from CSV Files
nodes_df = pd.read_csv(r'C:\Users\jasmi\Downloads\nodes2.csv')
edges_df = pd.read_csv(r'C:\Users\jasmi\Downloads\edges_weighted2.csv')

#Create a Graph
G = nx.from_pandas_edgelist(edges_df, 'source', 'target')

#Calculate Degree Distribution
degree_distribution = nx.degree_histogram(G)

# Print degree distribution and connected components
print("Degree Distribution:", degree_distribution)

# Plot degree distribution (optional)
plt.bar(range(len(degree_distribution)), degree_distribution)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution')
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

#Read Data from CSV Files
nodes_df = pd.read_csv(r'C:\Users\jasmi\Downloads\nodes2.csv')
edges_df = pd.read_csv(r'C:\Users\jasmi\Downloads\edges_weighted2.csv')

#Create a Graph
G = nx.from_pandas_edgelist(edges_df, 'source', 'target')

#Calculate Degree Distribution
degree_distribution = nx.degree_histogram(G)

# Plot degree distribution on log-log scale
plt.figure(figsize=(10, 6))
plt.loglog(range(len(degree_distribution)), degree_distribution, 'bo-')
plt.xlabel('Degree (log scale)')
plt.ylabel('Frequency (log scale)')
plt.title('Degree Distribution (log-log scale)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

#Read Data from CSV Files
nodes_df = pd.read_csv(r'C:\Users\jasmi\Downloads\nodes2.csv')
edges_df = pd.read_csv(r'C:\Users\jasmi\Downloads\edges_weighted2.csv')

#Create a Graph
G = nx.from_pandas_edgelist(edges_df, 'source', 'target')

#Calculate Degree Distribution
degree_distribution = nx.degree_histogram(G)
degrees = range(len(degree_distribution))

# Plot degree distribution on linear scale
plt.figure(figsize=(10, 6))
plt.plot(degrees, degree_distribution, 'bo-')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution (Linear scale)')
plt.grid(True)
plt.show()
```
