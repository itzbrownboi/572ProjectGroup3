```python
# Read the CSV file into a Pandas DataFrame
movies = pd.read_csv("imdb_movies_subset_300.csv")

# Extract the 'crew' column and create an array of actor names
actor_names = []

for crew_list in movies['crew']:
    if not pd.isna(crew_list):
        actor = crew_list.split(", ")[0::2]
        actor_names.extend(actor)

unique_actors = []
for item in actor_names:
    if item not in unique_actors:
        unique_actors.append(item)
        
unique_actors = [item for item in unique_actors if item != ""]

print(len(unique_actors))
print(unique_actors)
#type(actor_names)

```

```python
    
# Creating the edges.csv
unique_actors1 = unique_actors[0:30]

# Initialize a DataFrame for edges
edges = pd.DataFrame(columns=['source', 'target', 'weight'])

# Compare permutation pairs of all actors
for i in range(len(unique_actors) - 1):
    for j in range(i + 1, len(unique_actors)):
        source = unique_actors[i]
        target = unique_actors[j]
        #print(f"Pair: {source} - {target}")

        # For each crew list
        for crew_list in movies['crew']:
            # Check for NaN values
            if not pd.isna(crew_list):
                crew = crew_list.split(", ")[0::2]

                # If the 2 actors are in the same crew
                if source in crew and target in crew:
                    # Make a new edge
                    new_edge = pd.DataFrame({'source': [source], 'target': [target], 'weight': [1]})
                    edges = pd.concat([edges, new_edge], ignore_index=True)
                    print(f"Pair found: {source} - {target}")
                    


edges
```

```python
edges.to_csv('edges.csv', index=False)
```

```python
#determine WEIGHTS


# Create a copy of the DataFrame
edges_weighted = edges.copy()

# Identify duplicate rows
duplicates = edges_weighted.duplicated(subset=['source', 'target'], keep=False)

# Update weights for duplicate rows in the copy
edges_weighted['weight'] = edges_weighted.groupby(['source', 'target'])['weight'].transform('sum')

# Keep only one instance of each duplicate in the copy
edges_weighted = edges_weighted[~duplicates | edges_weighted.duplicated(subset=['source', 'target'], keep='first')]

# Reset index in the copy (optional)
edges_weighted = edges_weighted.reset_index(drop=True)

# Print the original and modified copies
print("Original DataFrame:")
print(edges)

print("\nModified DataFrame (edges_weighted):")
print(edges_weighted)
```

```python
edges_weighted.to_csv('edges_weighted.csv', index=False)
```

```python
# change from naems to IDs

# Assuming you have a CSV file with columns id and name
nodes_mapping_file = 'nodes300.csv'
nodes_df = pd.read_csv(nodes_mapping_file)

# Create a copy of the original DataFrame
edges_mapped_df = edges_weighted.copy()

# Create a mapping dictionary from names to IDs
name_to_id_mapping = dict(zip(nodes_df['actorName'], nodes_df['id']))

# Replace source and target columns with their corresponding IDs in the copied DataFrame
edges_mapped_df['source'] = edges_mapped_df['source'].map(name_to_id_mapping)
edges_mapped_df['target'] = edges_mapped_df['target'].map(name_to_id_mapping)

# Display the original and copied DataFrames
print("Original DataFrame:")
print(edges)

print("\nDataFrame with IDs:")
print(edges_mapped_df)
```
